```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
# load pima indians dataset
#1. Chun bi data:sinh vin vo google g tn file v load v
#my mnh
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
print(dataset.shape)
# split into input (X) and output (Y) variables
#2. Gii thch X,Y cc ch s trong[], th thay i 1 ch s
#khc xem sao. Cho v d
# X l cc ch s ca cc ct trong file csv
X = dataset[:,0:8]
# Y l ch s ca ct th 9(lable)
Y = dataset[:,8]
# create model
#3. Xy dng model. Gii thch cc dng lnh sau
# To model Sequential 
model = Sequential()
# Thm 1 lp Dense vi 12 n ron, 8 u vo, activation='relu' cho u ra khng m 0 n max(u vo)
model.add(Dense(12, input_dim=8, activation='relu'))
# Thm 1 lp Dense vi 8 n ron, activation='relu' cho u ra khng m 0 n max(u vo)
model.add(Dense(8, activation='relu'))
# Thm 1 lp Dense vi 1 n ron, activation='sigmoid' cho u ra 0 n 1
model.add(Dense(1, activation='sigmoid'))
# Compile model
#4. compile v gii thch 3 tham s loss, optimizer, metrics

# loss: hm mt mt, optimizer: thut ton ti u, metrics: nh gi

# binary_crossentropy l hm mt mt cho bi ton phn loi 2 lp
# adam l thut ton ti u
# accuracy l nh gi

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print (model.summary())
# Fit the model
#5. fit model: gii thch cc tham sepoch, batch
# verbose=0: khng hin th qu trnh train, verbose=1: hin th qu trnh train, verbose=2: hin th qu trnh train ngn gn
# validation_split: t l d liu dng  kim tra
# epochs: s ln hc li ton b d liu
# batch_size: s lng d liu c chia ra  train
history = model.fit(X, Y, validation_split=0.33, epochs=150, batch_size=10, verbose=0)
# list all data in history
#6. Gii thch history
# history.history l 1 dict cha cc gi tr loss, accuracy, val_loss, val_accuracy
print(history.history.keys())
# summarize history for accuracy
#7. Gii thch Biu 
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
#8. Gii thch biu 
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#Import required packages
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
# seed(2018) l hm khi to gi tr ngu nhin cho hm random
np.random.seed(2018) 
# random.random(1000,3) l hm to 1000 dng 3 ct vi cc gi tr ngu nhin t 0 n 1
# random.random((500, 3) l hm to 500 dng 3 ct vi cc gi tr ngu nhin t 0 n 1
train_data, test_data = np.random.random((1000, 3)), np.random.random((500, 3))
# random.randint(2, size=(1000, 1)) l hm to 1000 dng 1 ct vi cc gi tr ngu nhin t 0 n 2
labels = np.random.randint(2, size=(1000, 1))
# Sequential() l hm to model Sequential
model = Sequential()
# Dense(5, input_dim=3, activation='relu') l hm to 1 lp Dense vi 5 n ron, 3 u vo, activation='relu' cho u ra khng m 0 n max(u vo)
model.add(Dense(5, input_dim=3, activation='relu'))
model.add(Dense(4, activation='relu'))
# u ra model.add(Dense(1, activation='sigmoid')) l 1 n ron, activation='sigmoid' cho u ra 0 n 1
model.add(Dense(1, activation='sigmoid'))

# compile: hm bin dch model, loss: hm mt mt, optimizer: thut ton ti u, metrics: nh gi
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#Train the model and make predictions
history = model.fit(train_data, labels, epochs=10, batch_size=32)

import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.ylabel('accuracy')
plt.show()
#Make predictions from the trained model
predictions = model.predict(test_data)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
# Generate dummy training dataset
# np.random.seed(2018) l hm khi to gi tr ngu nhin cho hm random ti seed(2018)
# vi mi seed(value) th cc gi tr ngu nhin s ging nhau trong 1 phin lm vic
# seed(0) th cc gi tr ngu nhin s khc seed(1), seed(2) 
np.random.seed(2018)
x_train = np.random.random((6000,10))
y_train = np.random.randint(2, size=(6000, 1)) # random (0->1) tp nhn
# Generate dummy validation dataset
x_val = np.random.random((2000,10))
y_val = np.random.randint(2, size=(2000, 1))
# Generate dummy test dataset
x_test = np.random.random((2000,10))
y_test = np.random.randint(2, size=(2000, 1)) 
#Define the model architecture
model = Sequential()
# Dense: hm to 1 lp Dense vi 64 n ron, 10 u vo, activation='relu' cho u ra khng m 0 n max(u vo)
model.add(Dense(64, input_dim=10,activation = "relu")) #Layer 1
model.add(Dense(32,activation = "relu")) #Layer 2
model.add(Dense(16,activation = "relu")) #Layer 3
model.add(Dense(8,activation = "relu")) #Layer 4
model.add(Dense(4,activation = "relu")) #Layer 5
model.add(Dense(1,activation = "sigmoid")) #OutputLayer
#Configure the model
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
#Train the model
history = model.fit(x_train, y_train, batch_size=64, epochs=3, validation_data=(x_val,y_val))
#evaluate(x=None, y=None, batch_size=None, verbose=1,sample_weight=None, steps=None)
print(model.evaluate(x_test,y_test))
print(model.metrics_names)
#print 10 predictions
pred = model.predict(x_test)
pred[:10]


import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
